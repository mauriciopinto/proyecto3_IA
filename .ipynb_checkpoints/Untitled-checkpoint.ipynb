# kmeans

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
from math import sqrt
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
import seaborn as sns
```

```python
df
```

```python
df = df.reset_index()
```

```python
df= df.drop('index', 1)
```

```python
df
```

```python
ss= StandardScaler().fit_transform(df.drop('classes',1))
```

```python
ss
```

```python
pca = PCA(n_components=5)
comp = pca.fit_transform(df.drop('classes', 1))
```

```python
pca
```

```python
features = range(pca.n_components_)
plt.bar(features, pca.explained_variance_ratio_, color='blue')
plt.xlabel('PCA features')
plt.ylabel('variance %')
plt.xticks(features)
```

```python
dfcomponents = pd.DataFrame(comp)
```

```python
inertias = []

for k in range(1,10):
    km = KMeans(n_clusters=k).fit(dfcomponents)
    inertias.append(km.inertia_)
    
plt.plot(list(range(1,10)), inertias, '-o', color='black')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(list(range(1,10)))
plt.show()

```

```python
cols = {'col1','col2','col3','col4','col5'}
```

```python
dfcomponents.columns = cols
```

```python
dfcomponents
```

```python
km3 = KMeans(n_clusters=5).fit(dfcomponents)

dfcomponents['Labels'] = km3.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(dfcomponents['col1'], 
                dfcomponents['col2'], 
                #dfcomponents['col3'], 
                #dfcomponents['col4'], 
                #dfcomponents['col5'], 
                hue=dfcomponents['Labels'], 
                palette=sns.color_palette('hls', 5)
               )
plt.title('KMeans with 5 Clusters')
plt.show()
```

```python
fig = plt.figure(figsize=(10,8))

for col in cols:
    sns.swarmplot(x='Labels', y=col, data=dfcomponents)
    plt.show()
```

# ahc

```python
from sklearn.cluster import AgglomerativeClustering
```

```python
agglom = AgglomerativeClustering(n_clusters=5, linkage='average').fit(dfcomponents)

dfcomponents['Labels'] = agglom.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(dfcomponents['col1'],
                dfcomponents['col2'],
                hue=dfcomponents['Labels'], 
                palette=sns.color_palette('hls', 5))
plt.title('Agglomerative with 5 Clusters')
plt.show()
```

```python
from scipy.cluster import hierarchy 
from scipy.spatial import distance_matrix 

dist = distance_matrix(dfcomponents, dfcomponents)
print(dist)
```

```python
Z = hierarchy.linkage(dist, 'complete')
```

```python
plt.figure(figsize=(18, 50))
dendro = hierarchy.dendrogram(Z, leaf_rotation=0, leaf_font_size=12, orientation='right')
```

```python
Z = hierarchy.linkage(dist, 'average')
plt.figure(figsize=(18, 50))
dendro = hierarchy.dendrogram(Z, leaf_rotation=0, leaf_font_size =12, orientation = 'right')
```

# dbscan

```python
from sklearn.cluster import DBSCAN 

db = DBSCAN(eps=11, min_samples=6).fit(dfcomponents)

dfcomponents['Labels'] = db.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(dfcomponents['col1'], 
                dfcomponents['col2'],
                hue=dfcomponents['Labels'], 
                palette=sns.color_palette('hls', np.unique(db.labels_).shape[0]))
plt.title('DBSCAN epsilon11, min samples 6')
plt.show()
```

# mean shift

```python
from sklearn.cluster import MeanShift, estimate_bandwidth

bandwidth = estimate_bandwidth(dfcomponents, quantile=0.1)
ms = MeanShift(bandwidth).fit(dfcomponents)

dfcomponents['Labels'] = ms.labels_
plt.figure(figsize=(12, 8))
sns.scatterplot(dfcomponents['col1'],
                dfcomponents['col2'], 
                hue=dfcomponents['Labels'], 
                palette=sns.color_palette('hls', np.unique(ms.labels_).shape[0]))
plt.plot()
plt.title('MeanShift')
plt.show()
```
